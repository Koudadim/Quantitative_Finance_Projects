# Options pricing and Greeks analysis using Black-Scholes-Merton (BSM) Framework

The Black-Scholes model, developed by Fischer Black, Myron Scholes, and Robert Merton in 1973, revolutionized the field of financial derivatives by providing a closed-form solution for pricing European options. It assumes a continuous-time framework and is widely used in both academic research and practical applications.

## 1. The BSM model framework

### 1.1. Fundamental hypothesis

The Black–Scholes model assumes that the market consists of at least one risky asset, usually called the stock, and one riskless asset, usually called the money market, cash, or bond.

#### Asumptions about the assets:
* Risk-free rate $r$: The rate of return on the riskless asset is constant and thus called the risk-free interest rate
* Random walk: The instantaneous log return of the stock price is an infinitesimal random walk with drift; more precisely, the stock price $S_t$ at any time $t$ follows a geometric Brownian motion $W_t$, and it is assumed that the drift $\mu$ and volatility $\sigma$ of the motion are constant. Thus the underlying asset price dynamics is as follow:
$$ dS_t = \mu S_t dt + \sigma S_t dW_t $$
Using the Itô's formula, we obtain the solution to this Stochastique Differential Equation (SDE):
$$ S_t = S_0 exp\left( \sigma W_t + (\mu - \dfrac{\sigma ^2}{2})t\right) $$

Note that under the risk neutral probability, $\mu = r$ and then:

The dynamics of the risky asset is given by:
$$ dS_t = r S_t dt + \sigma S_t dW_t $$

and:

$$
S_t = S_0 exp\left( \sigma W_t + (r - \dfrac{\sigma ^2}{2})t\right)
$$

* The stock does not pay a dividend.

#### Assumptions about the market:

* No arbitrage opportunity (i.e., there is no way to make a riskless profit).
* Ability to borrow and lend any amount, even fractional, of cash at the riskless rate $r$.
* Ability to buy and sell any amount, even fractional, of the stock (this includes short selling).
* The above transactions do not incur any fees or costs (i.e., frictionless market).

Note that by extension, the BSM model can be used for the valuation of derivatives written on instruments paying continuous yield dividends. In that case, given $q$ a dividend yield, the dividend paid over the time period $[t,t+dt]$ is modelled by: $qS_t dt$.


### 1.2. Notations

The BSM model is used for the valuation of European options (Call and Put that can be exercised only at their maturity).

Notations:
* $V(S,t)$ is the price of the option as a function of the underlying asset S at time t, in particular:
* $C(S,t)$ is the price of a European call option and
* $P(S,t)$ is the price of a European put option.
* $T$ is the time of option expiration.
* $\tau =T-t$ is the time until maturity: 
* $K$ is the strike price of the option, also known as the exercise price.
* $D= e^{-r \tau }$ the discount factor; 
* $F= e^{r \tau }S=  \dfrac{S}{D} $ the forward price of the underlying asset

* $N$ denotes the standard normal cumulative distribution function:

$$
N(x)=\frac{1}{\sqrt{2 \pi}} \int_{-\infty}^{x} e^{-x^{2} / 2} d z . \quad N^{\prime}(x)=\frac{d N(x)}{d x}=\frac{1}{\sqrt{2 \pi}} e^{-x^{2} / 2}
$$


### 1.3. Risk-Neutral Valuation

The Black-Scholes model operates under the risk-neutral measure, where the expected return of the underlying asset is equal to the risk-free rate. Recall that under the risk-neutral measure, the discounted price process is a martingale.
This simplifies the pricing of derivatives by removing the need to estimate the asset's expected return.

#### For European options written on underlying assets that do not deliver dividend.

The Black-Scholes equation is given by:


$$
\frac{\partial V}{\partial t}+\frac{1}{2} \sigma^{2} S^{2} \frac{\partial^{2} V}{\partial S^{2}}+r S \frac{\partial V}{\partial S}-r V=0
$$

with the following terminal conditions:
* $V(x,T) = max(x - K, 0) $ if it is a European call option
* $V(x,T) = max(K - x, 0) $ if it is a European put option

The price of a European call option is given by:


$$
C\left(S_{t}, t\right)=N\left(d_{+}\right) S_{t}-N\left(d_{-}\right) K e^{-r(T-t)} \quad \text{ or } \quad C(F, \tau)=D\left[N\left(d_{+}\right) F-N\left(d_{-}\right) K\right]
$$

And by the Call-Put parity relationship, we deduce the European Put option:

$$
\begin{aligned}
P\left(S_{t}, t\right) & =K e^{-r(T-t)}-S_{t}+C\left(S_{t}, t\right) \\
& =N\left(-d_{-}\right) K e^{-r(T-t)}-N\left(-d_{+}\right) S_{t} 
\end{aligned}
$$

Which is equivalent to:

$$
\begin{aligned}
P(F, \tau)= & D\left[N\left(-d_{-}\right) K-N\left(-d_{+}\right) F\right]
\end{aligned}
$$

Where:

$$
\begin{aligned}
% d_{+}=d_{1} \text { and } d_{-}= & d_{2} \\
d_{+} & =\frac{1}{\sigma \sqrt{T-t}}\left[\ln \left(\frac{S_{t}}{K}\right)+\left(r+\frac{\sigma^{2}}{2}\right)(T-t)\right]  & =\frac{1}{\sigma \sqrt{\tau}}\left[\ln \left(\frac{F}{K}\right)+\frac{1}{2} \sigma^{2} \tau\right] \quad \text{ and } \quad
d_{-} = d_{+}-\sigma \sqrt{T-t} & =d_{+}-\sigma \sqrt{\tau}
\end{aligned}
$$

#### For European options written on underlying assets paying continuous yield dividends

The arbitrage-free price implied by the Black–Scholes model is:


$$
\begin{aligned}
& C\left(S_{t}, t\right)=e^{-r(T-t)}\left[F N\left(d_{1}\right)-K N\left(d_{2}\right)\right] \quad ; \quad P\left(S_{t}, t\right)=e^{-r(T-t)}\left[K N\left(-d_{2}\right)-F N\left(-d_{1}\right)\right]  \\ 
& F=S_{t} e^{(r-q)(T-t)} ; \quad d_{1}=\frac{1}{\sigma \sqrt{T-t}}\left[\ln \left(\frac{S_{t}}{K}\right)+\left(r-q+\frac{1}{2} \sigma^{2}\right)(T-t)\right] ; \quad d_{2}=d_{1}-\sigma \sqrt{T-t}
\end{aligned}
$$

#### 3.2.1. The Delta $\delta$

By definition, the delta $\delta$ is the first partial derivative of the premium with respect to $S$; it thus reflects the sensitivity of the option’s value to the underlying’s market price.

Using the call-put parity relation, for an option on a spot asset that does not pay a dividend:   
$$\delta_C- \delta_P= \delta_S=1$$
It follows from its mathematical definition that delta is the slope of the tangent to the curve representing the option price as a function of the underlying’s price.
For a small variation of the underlying price one has:

$$ \text{Variation in the option value } =\delta_{\text {option }} \times \text{  Variation in the underlying price}$$

If the European options are written on instruments paying a dividend yield $q$, then:
* $\delta_{call} = e^{-q (T-t)} N\left(d_{1}\right)$
* $\delta_{put} = - e^{-q (T-t)} N\left(d_{1}\right)$

##### Fundamental properties of the delta

* A call has a positive delta between 0 and 1. 
* A put has a negative delta between -1 and 0. 
* The underlying asset has a delta of 1; a risk-free asset has a delta of zero. 
* The absolute value of the option’s delta is approximately equal to the exercise probability.
* An option which is at-the-money often has the absolute value of its delta near 0.5 (intuitively, one thinks that it has approximately one chance in two of expiring in-the-money). 
* An option deep in-the-money has a delta of absolute value near to 1 (it is highly probable that it will be exercised). 
* An option which is deep out-of-the-money has a delta that is nearly zero (there is practically no chance of its being exercised).
* Moreover the $\delta$ is additive (as are all the Greeks), so that the $\delta$ of a portfolio is the weighted sum of the deltas of its components.

#### 3.2.2. The Gamma $\Gamma$

By definition, the gamma is the derivative of the delta, or the second derivative of the option’s value with respect to the underlying’s price.

Notice that the gamma of the underlying is zero and that by call-put parity (taking its derivative twice), we have: $\Gamma_C= \Gamma_P$, regardless of the pricing model employed.

If the European options are written on instruments paying a dividend yield $q$ then:
$$ \Gamma_{call}= \Gamma_{put} = e^{-q (T-t)} \dfrac{N^{\prime}(d_1)}{S\sigma \sqrt{T-t}} $$

The gamma can be viewed as the sensitivity of delta to variations in the underlying price; it is related to the curvature of the curve representing the option price as a function of S.

We note that the gamma of an option that is roughly at-the-money increases as it approaches expiration, while it decreases for options a long way in and out of the money.

Moreover the $\Gamma$ is additive (as are all the Greeks), so that the $\Gamma$ of a portfolio is the weighted sum of the vegas of its components.

#### 3.2.3. The Vega $v$ (not a Greek letter but used as so)

Given the fact that the underlying’s volatility $\sigma$ is, in practice, not constant or even deterministic, it is useful to compute the sensitivity of option values to a change in $\sigma$.

By definition, the vega ν is the partial derivative of the underlying’s premium with respect to $\sigma$. 
Due to the call-put parity, $ν_{call}= ν_{put}$ no matter which model is employed.
For given $S$, $\sigma$ and $(T–t)$, the vega and the gamma are proportional: 
$$ v = \sigma (T-t) S^2 \Gamma $$

However, ordinarily the first decreases with time while the second increases.

##### Some properties of the vega:
* The vega for the underlying is zero. 
* The vega of a standard option (call or put) is positive. 
* The vega of an option increases with its duration (like the square root of time), and therefore decreases with the passage of time (see figure below). 
* The vega of an option is maximum when it is close to being in-the-money (see see figure below). 
* The vega is additive (as are all the Greeks), so that the vega of a portfolio is the weighted sum of the vegas of its components.

In practice, it is essential to know this sensitivity; other things aside:
* A long position (positive vega) is a winner (resp. loser) when the implied volatility increases (resp. decreases); 
* On the contrary a short position (negative vega) is a winner (resp. loser) when the implied volatility decreases (resp. increases)
* A hedge is thus all the better, other things being equal, if the portfolio vega is near zero.

If the European options are written on instruments paying a dividend yield $q$ then:
$$ \text{Vega} = S_t e^{-q (T-t)} N^{\prime}(d_1) \sqrt{T-t} $$

#### 3.2.4. The theta $\theta$

The value of an option increases with its maturity, and thus its value decreases as time goes by, other things being unchanged.
By definition, the theta is the partial derivative of the premium with respect to time.

#### Properties concerning the other options’ thetas:
* Options in general have a negative theta; a long position in options (usually) has a negative theta, a short position a positive theta. 
* The underlying has no theta. 
* An option’s theta is maximum near at-the-money. 
* For options around at-the-money, theta increases (in absolute value) as the option nears its expiration. 
* As the theta is additive, a portfolio’s theta is the weighted sum of the thetas of its components.

If the European options are written on instruments paying a dividend yield $q$ then:

$$ 
\theta_{call} = -\frac{S_t N^{\prime}(d_1) \sigma e^{-q(T-t)}}{2\sqrt{(T-t)}} + q S_t e^{-q(T-t)} N(d_1) - r K e^{-r (T-t)} N(d_2) \quad ; \quad
\theta_{put} = \theta_{call} + r K e^{-r(T-t)} - q S_t e^{-q (T-t)}
$$

#### 3.2.5. The Rho $\rho$

The rho measures the influence of a variation in the interest rate r on the option’s value. By definition, the rho is the partial derivative of the premium with respect to the interest rate.

For a standard call and put whose underlying asset pays dividend rate $q$ or not, one has:
$$ \rho_{call} = K (T-t) e^{-r(T-t)} N(d_2) \quad \text{ and } \quad \rho_{put} = -K (T-t) e^{-r(T-t)} N(-d_2) $$

Note that the option replicating portfolio (that synthesizes the option) is made up of a zero-coupon bond whose maturity coincides with the option’s (a short position for a call and a long one for a put) and a quantity of the underlying equal to delta (a long position for a call, and short for a put). The risk reflected in the rho is that of a variation in the price of a zero-coupon bond; from this viewpoint, it is “symmetric” to the risk of variation in the underlying’s price linked to the delta.

#### 3.2.6. The Psi $\psi$

The Psi is a sensitivity measure in option pricing that evaluates the effect of changes in the dividend yield on the price of an option. It quantifies how the value of an option changes with respect to changes in the continuous dividend yield $q$ of the underlying asset.

Consider a dividend-paying stock with a call option. If the dividend yield increases by 1%, the psi value provides the corresponding decrease in the call option price. For a put option, it provides the corresponding increase.

By understanding psi, traders can effectively hedge against risks associated with changes in the dividend yield, making it an essential sensitivity measure in options pricing.

##### Properties
* For Call Options: A higher dividend yield reduces the attractiveness of holding the underlying asset, lowering the call option price. Hence, psi for a call option is typically negative.
* For Put Options: A higher dividend yield increases the attractiveness of selling the underlying asset, increasing the put option price. Hence, psi for a put option is typically positive.

$$ \psi_{call} = -(T-t) S_t e^{-q (T-t)} N(d_1) \quad \text{ and } \quad \psi_{put} = (T-t) S_t e^{-q(T-t)} N(-d_1)$$

# 4. The limitations of the BSM model

The Black-Scholes-Merton (BSM) model is a cornerstone of modern finance, widely used for pricing European options. However, it has several limitations, which arise due to the assumptions it makes and the complexities of real-world markets.

#### The ssumption of constant volatility
* The model assumes that the volatility ($\sigma$) of the underlying asset is constant over time.
* **Reality**: In practice, volatility changes due to market conditions, leading to the "volatility smile" or "volatility skew," which the BSM model cannot explain.

#### The assumption of constant risk-free rate
* The model assumes a constant risk-free interest rate ($r$) over the life of the option.
* **Reality**: Interest rates can fluctuate over time due to macroeconomic factors.

#### The no-dividends or simplistic dividend treatment
The original model assumes that the underlying asset pays no dividends. While there are modifications to account for dividends, the assumptions can be simplistic and may not capture complex dividend schedules.

#### The European-style options only
* The model is designed for European options, which can only be exercised at expiration.
* **Reality**: Many options traded in the real world, such as American options, can be exercised before maturity, requiring different models.

#### The assumption of lognormal price distribution for underlying asset
* The model assumes that the underlying asset price follows a lognormal distribution, implying that returns are normally distributed.
* **Reality**: Financial markets often exhibit "fat tails" and "skewness," leading to higher probabilities of extreme price movements.

#### The no-transaction costs or taxes
* The BSM model assumes frictionless markets, with no transaction costs, taxes, or liquidity issues.
* **Reality**: These factors can significantly impact trading strategies and option pricing.

#### The no arbitrage assumption
* The model assumes that markets are arbitrage-free, meaning there are no opportunities for riskless profit.
* **Reality**: Temporary arbitrage opportunities can exist due to market inefficiencies.

#### The continuous trading assumption
* The model assumes that trading occurs continuously, allowing perfect hedging at every moment.
* **Reality**: In practice, trading happens in discrete intervals, and perfect hedging is unattainable.

#### The single source of risk assumption
* The model assumes that the only source of uncertainty is the price of the underlying asset.
* **Reality**: Other risks, such as counterparty risk, liquidity risk, and operational risk, can also impact option pricing.

#### The assumption of infinite liquidity of the financial market
* The model assumes that markets are infinitely liquid, meaning any quantity of the underlying asset can be traded without affecting its price.
* **Reality**: Large trades can impact market prices, especially in less liquid markets.
